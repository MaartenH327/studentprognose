```python
import pandas as pd
import json
import numpy as np
from scripts.load_data import *
```

```python
configuration = load_configuration("../configuration/configuration.json")

paths = configuration["paths"]
data_student_numbers_f = pd.read_excel(paths["path_student_count_first-years"])
data_student_numbers_h = pd.read_excel(paths["path_student_count_higher-years"])
data_student_numbers_v = pd.read_excel(paths["path_student_volume"])
```

```python
starting_year = 2021
ending_year = 2022
predict_year = 2023
```

```python
combined_data_student_numbers = data_student_numbers_f.merge(data_student_numbers_h, on=["Croho groepeernaam", "Herkomst", "Collegejaar"])
```

```python
combined_data_student_numbers = combined_data_student_numbers.rename(columns={"Aantal_studenten_x": "Aantal_studenten_f", "Aantal_studenten_y": "Aantal_studenten_h"})
combined_data_student_numbers
```

```python
combined_data_student_numbers = combined_data_student_numbers[(combined_data_student_numbers["Collegejaar"] >= starting_year) & (combined_data_student_numbers["Collegejaar"] <= ending_year)]
```

```python
combined_data_student_numbers["Ratio"] = combined_data_student_numbers["Aantal_studenten_h"].divide(combined_data_student_numbers["Aantal_studenten_f"])
combined_data_student_numbers["Ratio"] = combined_data_student_numbers['Ratio'].replace(np.inf, np.nan)
combined_data_student_numbers[combined_data_student_numbers["Croho groepeernaam"] == "M Geneeskunde"]
```

```python
ratio = combined_data_student_numbers[["Croho groepeernaam", "Herkomst", "Ratio"]].groupby(['Croho groepeernaam', 'Herkomst'], as_index=False).mean()
ratio
```

```python
ratio[ratio["Croho groepeernaam"] == "M Geneeskunde"]
```

```python
total_mae = 0.0
total_mape = 0.0
count = 0.0
count_mape = 0.0

for _, row in ratio.iterrows():
    programme = row["Croho groepeernaam"]
    origin = row["Herkomst"]
    first_year_row = data_student_numbers_f[(data_student_numbers_f["Croho groepeernaam"] == programme) &
                                                    (data_student_numbers_f["Herkomst"] == origin) &
                                                    (data_student_numbers_f["Collegejaar"] == predict_year)]
    higher_year_row = data_student_numbers_h[(data_student_numbers_h["Croho groepeernaam"] == programme) &
                                                    (data_student_numbers_h["Herkomst"] == origin) &
                                                    (data_student_numbers_h["Collegejaar"] == predict_year)]
    
    # print(first_year_row["Aantal_studenten"])
    # print(higher_year_row["Aantal_studenten"])

    if len(first_year_row) > 0 and len(higher_year_row) > 0:
        f_s = first_year_row["Aantal_studenten"].iloc[0]
        h_s = higher_year_row["Aantal_studenten"].iloc[0]
        r = row["Ratio"]

        p = f_s * r
        mae = abs(h_s - p)
        mape = abs((h_s - p) / h_s)
        print(f"{programme}, {origin}: {mae} & {mape}")
        total_mae += mae
        count += 1.0
        
        if h_s >= 10:
            total_mape += mape
            count_mape += 1.0

print(f"Final MAE: {total_mae / count}")
print(f"Final MAPE: {total_mape / count_mape}")

print(f"Total MAE: {total_mae}")
print(f"Total MAPE: {total_mape}")
print(f"Count: {count} & {count_mape}")
```
